# Assignment Random Forest(fraud)

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier as RF
from sklearn import tree
from sklearn.metrics import classification_report
```

```python
fraud = pd.read_csv("E:DATA SETS/Fraud_check.csv")
```

```python
fraud
```

```python
fraud.isnull().sum()
```

```python
fraud1 = fraud
```

```python
rename = {'Marital.Status' : 'marital_status',
         'Taxable.Income' : 'taxable_income',
         'City.Population' : 'city_population',
         'Work.Experience' : 'work_experience'}
fraud1.rename(columns=rename, inplace= True )
```

```python
fraud1["tax_status"] = np.where(fraud1['taxable_income']<=30000,'Risky','Good') 
```

```python
fraud1
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
fraud1['Undergrad']= label_encoder.fit_transform(fraud1['Undergrad']) 
fraud1['marital_status']= label_encoder.fit_transform(fraud1['marital_status']) 
fraud1['Urban']= label_encoder.fit_transform(fraud1['Urban'])
```

```python
fraud1
```

```python
fraud1.drop(['taxable_income'], axis=1, inplace= True)
fraud1
```

```python
X=fraud1.iloc[:,:5]
X
```

```python
fraud1.tax_status.value_counts()
```

```python
Y=fraud1.iloc[:,5]
Y
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=40,test_size=0.2)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
model = RF(n_jobs = 3,n_estimators = 15, oob_score = True, criterion = "entropy")
model.fit(X_train,Y_train)
```

```python
model.estimators_
```

```python
model.n_features_
```

```python
model.oob_score_
```

```python
prediction = model.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(Y_train,prediction)
```

```python

np.mean(prediction == Y_train)
```

```python
pred_test = model.predict(X_test)
```

```python
acc_test =accuracy_score(Y_test,pred_test)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier as RF
from sklearn import tree
from sklearn.metrics import classification_report
```

```python
data = pd.read_csv("E:DATA SETS/Company_Data.csv")
```

```python
data
```

```python
fraud.isnull().sum()
```

```python
data1 = data.copy() 
```

```python
data1 = pd.get_dummies(data1, columns={'ShelveLoc','Urban','US'})
```

```python
data1
```

```python
data1["Sale"] = np.where(data1['Sales']<=7.5,'Bad','Good') 
```

```python
data1
```

```python
X=data1.iloc[:,:-1]
X
```

```python
Y = data1.Sale
```

```python
Y
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=40,test_size=0.2)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
model = RF(n_jobs = 3,n_estimators = 15, oob_score = True, criterion = "entropy")
model.fit(X_train,Y_train)
```

```python
model.estimators_
```

```python
model.n_features_
```

```python
model.oob_score_
```

```python
prediction = model.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(Y_train,prediction)
```

```python

np.mean(prediction == Y_train)
```

```python
pred_test = model.predict(X_test)
pred_test
```

```python
acc_test =accuracy_score(Y_test,pred_test)
```
